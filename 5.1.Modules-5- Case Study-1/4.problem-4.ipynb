```python
import numpy, pandas 
import matplotlib.pyplot as plt 


df = pandas.read_csv('HollywoodMovies.csv')
```

criticsâ€™ evaluation of a movie and itsacceptance by the public  

```python
plt.scatter(df["RottenTomatoes"], df["AudienceScore"])
plt.xlabel("RottenTomatoes")
plt.ylabel("AudienceScore")
plt.title("RottenTomatoes vs AudienceScore")
plt.show()
```

RottenTomatoes vs Profitability

```python
plt.scatter(df["RottenTomatoes"], df["Profitability"])
plt.xlabel("RottenTomatoes")
plt.ylabel("Profitability")
plt.title("RottenTomatoes vs Profitability")
plt.ylim(0,2000)
plt.show()
```
