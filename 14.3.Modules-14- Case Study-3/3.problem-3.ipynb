```python
import numpy, pandas 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

df = pandas.read_csv('letterCG.data' , sep=' ')

df.head()
```

```python
df.drop(["Unnamed: 5", "Unnamed: 18", "yegvx"], axis=1, inplace=True)

df.fillna(0)

X = df.iloc[:, 1:]

Y = df["Class"]

train_x, test_x, train_y, test_y = train_test_split(X, Y, random_state=5, test_size=0.30)
```

```python
base_class = DecisionTreeClassifier(max_depth=2)

ada_Boost = AdaBoostClassifier(base_estimator=base_class, n_estimators=400, learning_rate=1, algorithm='SAMME')

ada_Boost.fit(train_x, train_y)

predicted = ada_Boost.predict(test_x)

metrics.accuracy_score(predicted, test_y)
```
