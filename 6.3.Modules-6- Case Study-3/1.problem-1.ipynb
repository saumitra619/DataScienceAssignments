```python
import numpy, pandas 
import matplotlib.pyplot as plt 
import seaborn as sns

df = pandas.read_csv('FyntraCustomerData.csv')

df.head()
```

```python
df_Time_web_Amount = df.filter(["Time_on_Website", "Yearly_Amount_Spent"])

df_Time_web_Amount.head()

sns.set(style="white", color_codes=True)
sns.jointplot(x="Time_on_Website",y="Yearly_Amount_Spent",  kind = "scatter",data=df_Time_web_Amount)
```

```python
#strong correlation
coff_Time_Amount = df_Time_web_Amount.corr()

sns.heatmap(coff_Time_Amount)
```
