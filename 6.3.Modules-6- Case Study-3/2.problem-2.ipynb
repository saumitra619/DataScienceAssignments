```python
import numpy, pandas 
import matplotlib.pyplot as plt 
import seaborn as sns

df = pandas.read_csv('FyntraCustomerData.csv')

df.head()
```

```python


df_Time_app_Amount = df.filter(["Time_on_App", "Yearly_Amount_Spent"])

sns.set(style="white", color_codes=True)
sns.jointplot(x="Time_on_App",y="Yearly_Amount_Spent",  kind = "scatter",data=df_Time_app_Amount)
```

```python
#strong correlation
coff_Time_web_Amount = df_Time_app_Amount.corr()
sns.heatmap(coff_Time_web_Amount)
```
